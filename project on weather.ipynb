```python
.head()
it shows the first N rows inthe data (by default N=5) 
data.head()

.shape()
it shows the total rows and no. of columns of the dataframe
data.shape()

.index    
this attribute provides the indes of the dataframe
data.index

.columns
shows the name of each column
data.columns

.dtypes
it shows the data-type of each column
data.dtypes

.unique()
In a column it shows all the unique values it can be applied on a single column
data['weatherhistory']. unique()

.nunique()
In a column it shows all the unique values it can be applied on a single column
data.nunique()

.count
it show the total no of non-null values in each column. it is applied on a single column
as well as the whole dataframe
data.count()

.value_counts
in a column, it shows all the unique vakus with their count. it can be applied
on a single column only
data['weatherhistory'].value_counts()

.info()
provides basic information about the dataframe
data.info()
```
